```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib
matplotlib.__version__
```

```python
# プロット対象ダミーデータの作成(OK/NGとで、データ数に偏りあり)
ok_data = np.random.normal(0,10,5000)
ng_data = np.random.normal(20,10,100)
```

### 縦軸/横軸共通(並べて描画)

```python
plt.figure()

plt.hist([ok_data,ng_data],bins=30,rwidth=0.8,label=['OK','NG'])

plt.xlabel('x value')
plt.ylabel('y value')
plt.grid(which='both')
plt.legend()

plt.show()
```

### 縦軸は別/横軸共通(並べて描画)

```python
def plot_hist_anomalydata_tile(ok_data,ng_data,bin_num,hist_range,title,is_save=False,save_fname=None):
    islog = False

    plt.figure()
    fig,ax1 = plt.subplots()
    ax2 = ax1.twinx()

    ax1.hist(ok_data,bins=bin_num,range=hist_range,align='mid',rwidth=0.4,log=islog,color='tab:blue',label='OK')
    ax2.hist(ng_data,bins=bin_num,range=hist_range,align='right',rwidth=0.4,log=islog,color='tab:orange',label='NG')

    ax1.set_xlabel('Score')
    ax1.set_ylabel('OK')
    ax2.set_ylabel('NG')

    ax1.grid(b=True,which='both')
    ax2.grid(b=False,which='both')

    handler1,label1 = ax1.get_legend_handles_labels()
    handler2,label2 = ax2.get_legend_handles_labels()
    ax1.legend(handler1+handler2,label1+label2,loc='best')

    plt.title(title)

    if is_save:
        plt.savefig(save_fname)
    #plt.show()

bin_num = 30
hist_range=(-40,50)

plot_hist_anomalydata_tile(ok_data,ng_data,bin_num,hist_range,title='plot_title')
```

### 縦軸は別/横軸共通(重ねて描画)

```python
def plot_hist_anomalydata_overlay(ok_data,ng_data,bin_num,hist_range,title,is_save=False,save_fname=None):
    islog = False

    plt.figure()
    fig,ax1 = plt.subplots()
    ax2 = ax1.twinx()

    ax1.hist(ok_data,bins=bin_num,range=hist_range,alpha=0.5,log=islog,color='tab:blue',label='OK')
    ax2.hist(ng_data,bins=bin_num,range=hist_range,alpha=0.5,log=islog,color='tab:orange',label='NG')

    ax1.set_xlabel('Score')
    ax1.set_ylabel('OK')
    ax2.set_ylabel('NG')

    ax1.grid(b=False,which='both')
    ax2.grid(b=False,which='both')

    handler1,label1 = ax1.get_legend_handles_labels()
    handler2,label2 = ax2.get_legend_handles_labels()
    ax1.legend(handler1+handler2,label1+label2,loc='best')

    plt.title(title)

    if is_save:
        plt.savefig(save_fname)
    #plt.show()

bin_num = 30
hist_range=(-40,50)

plot_hist_anomalydata_overlay(ok_data,ng_data,bin_num,hist_range,title='plot_title')
```
