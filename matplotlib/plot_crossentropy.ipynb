```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func_cross_entropy(p,q):
  return( -p * np.log(q) - (1-p) * np.log(1-q) )
```

```python
x = np.arange( 0.01, 1.0, 0.01 )
y = np.arange( 0.01, 1.0, 0.01 )

X,Y = np.meshgrid(x,y)
Z = func_cross_entropy( X, Y )

fig = plt.figure()

from mpl_toolkits.mplot3d import Axes3D
ax = Axes3D(fig)

ax.set_xlabel('p')
ax.set_ylabel('q')
ax.set_zlabel('cross entropy(p,q)')

ax.plot_wireframe(X, Y, Z)
plt.show()
```
