```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func_gaussian_kernel(x):
  a = 1.0
  return -np.exp(a*x**2)
```

```python
x = np.arange(-1,1,0.1)
y = func_gaussian_kernel(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(which='both')
plt.title("Gaussian Kernel")
plt.show()
```
