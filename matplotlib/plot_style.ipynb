```python
# 線の種類
x = np.linspace(0,1,100)

plt.plot(x,np.sin(x),linestyle='-',label='solid')
plt.plot(x,np.sin(x)+0.1,linestyle='--',label='dashed')
plt.plot(x,np.sin(x)+0.2,linestyle=':',label='dotted')
plt.plot(x,np.sin(x)+0.3,linestyle='-.',label='dashdot')
plt.legend()
plt.show()
```
