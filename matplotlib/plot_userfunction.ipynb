```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ユーザー定義関数
def func(x):
    return 1.0 / (1.0 - x)

LABEL_STRING = r'$\frac{1}{1-x}$'
```

```python
# 定義域
X_MIN = -5.0
X_MAX = 7.0

Y_MIN = -6.1
Y_MAX = 6.1
```

```python
x = np.arange(X_MIN,X_MAX,0.01)
y_origin = func(x)
y = y_origin.copy()
y[y_origin > Y_MAX] = np.nan

plt.plot(x,y,label=LABEL_STRING)
plt.xlim([X_MIN,X_MAX])
plt.ylim([Y_MIN,Y_MAX])
plt.xlabel('x')
plt.ylabel('y')
plt.grid(which='both')
plt.legend()
plt.savefig('./output.svg')
plt.show()
```
