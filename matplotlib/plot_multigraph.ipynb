```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.rand(20)
time = np.linspace(0,10,20)
```

```python
plt.plot(time,data)
plt.show()
```

```python
plt.figure(figsize=(10.0,5.0))

plt.subplot(211)
plt.plot(time,data,label='test1')
plt.ylabel('test1')
plt.grid(which='both')
plt.legend()

plt.subplot(212)
plt.plot(time,data,label='test2')
plt.xlabel('time')
plt.ylabel('test2')
plt.grid(which='both')
plt.legend()

plt.show()
```
